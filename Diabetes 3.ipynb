we can see that for the features Glucose,BloodPressure,Insulin,BMI we have 0 as the min value .. these features can not have 0 as the min value..

```python
df['BMI'].replace(0,np.nan,inplace=True)
df['Glucose'].replace(0,np.nan,inplace=True)
df['BloodPressure'].replace(0,np.nan,inplace=True)
df['Insulin'].replace(0,np.nan,inplace=True)
df['SkinThickness'].replace(0,np.nan,inplace=True)
#missing value 
```

```python
pd.isnull(df).sum()
```

```python
df.head()
```

```python
#replace missing value in the mean
```

```python
newdata= df.fillna(df.mean()) 
```
